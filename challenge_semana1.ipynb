```python
import pandas as pd
from pandas import json_normalize
import numpy as np
```

```python
df = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
df.head()
```

```python
# Normalizando as colunas no formado json
custmer_normalize = json_normalize(df.customer)
phone_normalize = json_normalize(df.phone)
intrnet_normalize = json_normalize(df.internet)
account_normalize = json_normalize(df.account)


# Adicionando as colunas ao df antigo
dados = pd.concat([df, custmer_normalize, phone_normalize, intrnet_normalize, account_normalize], axis=1)
dados = dados.drop(columns=['customer', 'phone', 'internet', 'account'])
dados.head()
```

```python
# Avaliando os tipos dos dados
dados.info()
```

```python
dados['Charges.Total']
```

```python
# Vereficando valores vazios
dados.replace('', np.nan, inplace=True)
dados.replace(' ', np.nan, inplace=True)
dados.isnull().sum()
```

```python
# São 244 observações vazias para o valor de Churn. Verificando esses valores

total_churns_vazios = len(dados[dados['Charges.Total'].isnull()])
total_dados = len(dados)
print('Os churns vazios representam', (total_churns_vazios/total_dados)*100, '%')
```

```python
# Como os churns são 0.15% do todo, decidi retirar esses valores vazios para não prejudicar as análises
dados_v1 = dados.dropna()

print('Total de dados apos tratamentos:', dados_v1.shape)
print('Total de dados anterior aos tratamentos:', dados.shape)
```

```python
# Renomenado as colunas
dados_v1.rename(
    columns= {
        'customerID' : 'id_cliente',
        'Churn': 'cancelamento',
        'gender': 'genero',
        'SeniorCitizen' : 'idade_maior_65anos',
        'Partner' : 'tem_parceirx',
        'Dependents': 'tem_dependentes',
        'tenure': 'meses_de_contrato',
        'PhoneService': 'tem_assinatura_telefone',
        'MultipleLines': 'tem_linhas_telefone',
        'InternetService': 'assinatura_provedor_internet',
        'OnlineSecurity': 'assinatura_seguranca_online',
        'OnlineBackup': 'assinatura_backup',
        'DeviceProtection': 'assinatura_protecao_dispositivo',
        'TechSupport': 'assinatutra_suporte_tecnico',
        'StreamingTV': 'assinatura_tv_cabo',
        'StreamingMovies': 'assinatura_streaming',
        'Contract': 'tipo_contrato',
        'PaperlessBilling': 'preferencia_fatura_online',
        'PaymentMethod': 'forma_pagamento',
        'Charges.Monthly': 'total_mensal',
        'Charges.Total': 'total_gasto'
    },
            inplace=True)
print(dados_v1.columns)
```

```python
# Criando variável de gasto diário

dados_v1['gasto_diario'] = dados_v1['total_mensal']/30
dados_v1.head()
```

```python
# Transformando dados categoricos 'Yes' ou 'No' em contínuos 1 ou 0.
dados_v2 = dados_v1.replace(['Yes'], 1).replace(['No'], 0)

# Salvando os dados limpos
dados_v2.to_json('df_clean.json')
```
